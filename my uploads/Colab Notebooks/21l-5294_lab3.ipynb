```python
def reverseFile(input, output):
  with open(input, 'r') as inputFile:
    content = inputFile.read()
    reversedContent = content[::-1]
    with open(output, 'w') as outputFile:
        outputFile.write(reversedContent)
  print("File reversed")


input = '/content/alphabets.txt'
output = '/content/output.txt'

reverseFile(input, output)
```

```python
from collections import deque

def read_cube_from_file(filename):
    with open(filename, 'r') as file:
        cube = [list(line.strip().split()) for line in file]
    return cube

def dfs(cube):
    stack = [((0, 0), [(0, 0)])]
    visited = set([(0, 0)])

    while stack:
        current, path = stack.pop()
        x, y = current

        if cube[x][y] == 'G':
            return path

        neighbors = get_neighbors(x, y, cube)
        for neighbor in neighbors:
            if neighbor not in visited and cube[neighbor[0]][neighbor[1]] != '1':
                stack.append((neighbor, path + [neighbor]))
                visited.add(neighbor)

    return -1

def get_neighbors(x, y, cube):
    neighbors = []
    directions = [(1, 0), (-1, 0), (0, 1), (0, -1)]

    for dx, dy in directions:
        new_x, new_y = x + dx, y + dy
        if 0 <= new_x < len(cube) and 0 <= new_y < len(cube[0]):
            neighbors.append((new_x, new_y))

    return neighbors

def display_path(path):
    if path == -1:
        print("No path exists.")
    else:
        print("Path:", end=" ")
        for step in path:
            print(step, end=" ")
        print()

# Example usage:
filename = '/content/input.txt'
cube = read_cube_from_file(filename)

dfs_path = dfs(cube)
display_path(dfs_path)
```
