## Rectangle Class

```python
class Rectangle:
  def __init__(self,length,width):
    self.length = length
    self.width = width

  def area(self):
    return self.length*self.width
  def perimeter(self):
    return 2*(self.length+self.width)
  def diagonal_length(self):
    return (self.length**2 + self.width**2)**0.5

  def aspect_ratio(self):
    a = self.length
    b = self.width

    while b:
      a, b = b, a %b

    return f"{self.length/a}, {self.width/a}"

rec1 = Rectangle(10,8)
print(f"Area: {rec1.area()}")
print(f"Perimeter: {rec1.perimeter()}")
print(f"Diagonal Length: {rec1.diagonal_length()}")
print(f"Aspect Ratio: {rec1.aspect_ratio()}")
```

```python
def f(x):
  return x**2 - 4*x +5

def HCS_min(f,x0):
  x=x0

  while 0<x<5:
    neighbors = x-0.1,x+0.1
    best_neighbor = min(neighbors,key=f)
    if f(best_neighbor) >= f(x):
      return round(x),f(x)
    x = best_neighbor
  return round(x),f(x)

def HCS_max(f,x0):
  x=x0

  while 0<x<5:
    neighbors = x-0.1,x+0.1
    best_neighbor = max(neighbors,key=f)
    if f(best_neighbor) <= f(x):
      return round(x),f(x)
    x = best_neighbor
  return round(x),f(x)

print(f"x: {HCS_min(f,3.5)[0]} f(x):{HCS_min(f,3.5)[1]} ")
print(f"x: {HCS_max(f,1.5)[0]} f(x):{HCS_max(f,1.5)[1]} ")

```
